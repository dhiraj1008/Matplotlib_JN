## Scatter Plots
### - Scatter plots are great when your gone show relationship b/w two set of values(variables) and see if they are corelated
### In this MatplotLib Tutorial Part-7, we will be learning how to create scatter plots in Matplotlib. Scatter plots are great for determining whether two sets of data are correlated. If there is a correlation, scatter plots allow us to spot these trends. Let's get started... 


### Import Libraries

```python
import matplotlib.pyplot as plt
import random
```

```python

#plt.style.use('seaborn')

x=[5,7,6,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y=[7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
colors = [7,5,8,7,6,5,4,3,8,5,6,7,8,8,8,9,6,1,4,3]
sizes = [random.randint(200,500) for x in range(20)]

plt.scatter(x,y,c=colors,s=sizes,cmap='Greens',edgecolors='black',linewidth=1,alpha=0.75)

cbar = plt.colorbar()
# here color indicates the satisfication of the customers 
cbar.set_label("Satisfaction factor")

```

### Dealing with data from the csv file 
### this data represents the top most viewwd , liked  and like/dislike ratio video on youtube(got from youtube API)

```python

import pandas as pd

data = pd.read_csv("2019-05-31-data(youtube).csv")

view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count,likes,cmap='summer',c=ratio,edgecolor='black',linewidths=1,alpha=0.75)

cbar = plt.colorbar()
cbar.set_label("Like/Diske Ratio")
plt.title("Trending YouTube Videos")
plt.xlabel("view counts")
plt.ylabel("likes")
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()

plt.show()
```
