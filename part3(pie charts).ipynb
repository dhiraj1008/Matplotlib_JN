## Pie Chart
###           - pie charts are propotion of each category contributes to the whole..

### In this Python Programming video, we will be learning how to create pie charts in Matplotlib. Pie charts are great for visualizing your data in a way that clearly shows the proportions of each category. We'll learn how to plot a pie chart, customize how it looks, and more. Let's get started...

### chart=pie,category=slice in pie



### Import library

```python
import matplotlib.pyplot as plt
```

```python

# I draw pie charts when there are lesser portions (less than 7)..  

slices = [56789,55466,47654,36789,35432]
labels = ['Javascript','HTML/CSS','SQL','Python','Java']
explode = [0,0,0,0.1,0]#from what radius slice needs to be apart
plt.style.use('fivethirtyeight')
plt.pie(slices,labels=labels,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'},autopct="%1.1f%%")
#we can also specify the colors in list for specific slice by use color keyword argument(color=list)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.savefig("fig3(pie_chart).png")
plt.show()
# we can do lot more to customize the pie chart 
```
