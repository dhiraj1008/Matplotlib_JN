## Filling Area on Line Plots
###  fills give us some usefull insights  like we can use conditional  fills that will show that what area of lineplot fall above and below of certain threshold 
### In this Matplotlib Tutorial Part-5, we will be learning how to fill between our line plots in Matplotlib. Not only can this make our graphs look more professional, but we can also add useful information by filling areas based on specific thresholds. Let's get started...

### Import Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

### Import Data from csv file.

```python
data = pd.read_csv("Age,All_Devs,Python,JavaScript.csv")
```

```python
ages = data['Age']
dev_salaries=data['All_Devs']
py_salaries=data['Python']
js_salaries=data['JavaScript']

plt.plot(ages,dev_salaries,color='#444444',linestyle='--',label="ALL Devs")


plt.plot(ages,py_salaries,label="Python",color='y')
# overall median of dev_salary 
median = data['Python'].median() # median = 84016.0

plt.fill_between(ages,py_salaries,dev_salaries,
                 where = (py_salaries > dev_salaries),alpha=0.25,interpolate=True,label="Above Avg")

plt.fill_between(ages,py_salaries,dev_salaries,label="Below Avg",
                 where = (py_salaries < dev_salaries),alpha=0.25,interpolate=True)
plt.legend()

plt.title("Median Salary (USD) by Age")
plt.xlabel("Ages")
plt.ylabel("Median salary (USD)")
median
```
