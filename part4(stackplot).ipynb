## Stack Plots

### - Stack plots are generated by ploting different datasets vertically on top of another rather than overlapping with each one another.
### In this Python Programming video, we will be learning how to create stack plots in Matplotlib. Sometimes these are called Area Charts. These are similar to pie charts, but instead of showing us the proportions at a single instant, these allow us to see the proportions over a series of points. Let's get started...
### if we want to show parts of contribution in total contribution then use stack plot

###  Import Library

```python
import matplotlib.pyplot as plt
```

```python

minutes = [1,2,3,4,5,6,7,8,9]
#score of each player in total score at each minute
player1 = [8,6,5,5,4,2,1,1,0]
player2 = [0,1,2,2,2,4,4,4,4]
player3 = [0,1,1,1,2,2,3,3,4]

plt.style.use('fivethirtyeight')

labels = ['player1','player2','player3']
#adding custom colors
colors = ['#6d904f','#fc4f30','#008fd5'] # using hexacode.formate is #RRGGBB

# to draw a stack plot
plt.stackplot(minutes,player1,player2,player3,labels=labels,colors=colors)
#plt.legend(loc='upper left')
plt.legend(loc=(0.07,0.05))#specifying coordinates
plt.title("Stack Plot")
plt.savefig("fig4(stack_plot).png")
plt.tight_layout()
plt.show()
```
