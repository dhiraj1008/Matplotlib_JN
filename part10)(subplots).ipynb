## pyplot vs object oriented interface
# pyplot => matplotlib on the surface is made to imitate MATLAb's method  of generating plots,which is called pyplot.This is a state based interface
# OO Interface =>  when using  the oo interface,it helps to know how the matplotlib structures its plots .The final plot that we see as the output is a 'Figure'  object
# This way(creating our instance) is called stateful(plt => it has a current state in terms of ehat figures  we are currently working with and what axes we are  currently working with )

```python
import pandas as pd
from matplotlib import pyplot as plt 

# instead of using pyplot instance we are creating the new insatnce of figure and axes
plt.style.use("seaborn")

data = pd.read_csv("Age,All_Devs,Python,JavaScript.csv")
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data["Python"]
js_salaries = data['JavaScript']

fig,ax = plt.subplots() # Create a figure and a set of subplots.

ax.plot(ages,dev_salaries,color='red',linestyle='-',label="ALL Devs")
ax.plot(ages,py_salaries,label="Python",color='y')
ax.plot(ages,js_salaries,label="Java Script")

ax.legend()
ax.set_title("Median Salary (USD) by Age")
ax.set_xlabel("Age")
ax.set_ylabel("Median Salaru (USD)")
fig.savefig("fig10a).png")
#here only fig has access to the graph window not plt(because we are creating our new instatnce of plot)

plt.tight_layout()
plt.show()

#print(ax) # if no arguments are passed the subplots() by default its a single axis(1row ,1col) subplots Axes(0.113472,0.10596;0.867778x0.829798)
#print(ax) # if arguments are passed to the subplots() then [<Axes: > <Axes: >] two axes are returned we can unpack it in two variables as individual axes
```

# Create a figure and a set of subplots.

```python
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1,sharex=True) 

ax1.plot(ages,dev_salaries,color='#444444',linestyle='-',label="ALL Devs")
ax2.plot(ages,py_salaries,label="Python",color='y')
ax2.plot(ages,js_salaries,label="Java Script")

ax1.legend()
ax1.set_title("Median Salary (USD) by Age")
#ax1.set_xlabel("Age")
ax1.set_ylabel("Median Salaru (USD)")

ax2.legend()
#ax2.set_xlabel("Age")
ax2.set_ylabel("Median Salaru (USD)")

fig.savefig("fig10b).png")
#here only fig has access to the graph window not plt(because we are creating our new instatnce of plot)

plt.tight_layout()
plt.show()

#print(ax) # if no arguments are passed the subplots() by default its a single axis(1row ,1col) subplots Axes(0.113472,0.10596;0.867778x0.829798)
#print(ax) # if arguments are passed to the subplots() then [<Axes: > <Axes: >] two axes are returned we can unpack it in two variables as individual axes
```

# Plot on two different figures


```python

fig1,ax1 = plt.subplots() # Create a figure and a set of subplots.
fig2,ax2 = plt.subplots() 
ax1.plot(ages,dev_salaries,color='red',linestyle='-',label="ALL Devs")
ax2.plot(ages,py_salaries,label="Python",color='y')
ax2.plot(ages,js_salaries,label="Java Script")

ax1.legend()
ax1.set_title("Median Salary (USD) by Age")
ax1.set_xlabel("Age")
ax1.set_ylabel("Median Salaru (USD)")

ax2.legend()
ax2.set_title("Median Salary (USD) by Age")
ax2.set_xlabel("Age")
ax2.set_ylabel("Median Salaru (USD)")

fig1.savefig("fig10c).png")
fig1.savefig("fig10d).png")
```
